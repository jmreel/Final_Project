```python
import pandas as pd
import os
```

# Creating District dataframe

```python
# Creating dataframe for the General worksheet
general_df = pd.read_csv('General.csv')
general_df.head(10)
```

```python
# checking for na
general_df.isna().sum()
```

```python
# filtering for district
district_general_df = general_df.loc[general_df['Type'] == "District"]
```

```python
district_general_df.head(10)
```

```python
district_general_df.isna().sum()
```

```python
# dropping na with anything over 750
district_general_df = district_general_df.dropna(thresh=750, axis=1)
```

```python
list(district_general_df.columns)
```

```python
district_general_df.head()
```

```python
district_filtered_df = district_general_df[["RCDTS", 
    "Type",  
    "District", 
    "City", 
    "County",
    "District Type", 
    "District Size", 
    "Student Enrollment - Total",
    "Student Enrollment - Low Income %",
    "Avg Class Size – All Grades",
    "Avg Teaching Exp",
    "Bachelor Degree",
    "Masters Degree",
    "Teacher Retention Rate",
    "Teacher Avg Salary",
    "Admin Avg Salary"]]
district_filtered_df.head(20)
```

```python
district_filtered_df.count()
```
